```python
import pandas as pd
```

```python
csv1 = pd.read_csv("debt.csv")
```

```python
# now we will use replace
```

```python
# data.replace(to_replace=value to be replace,value=value which will replace)
```

```python
csv1.replace(to_replace=26,value=167)  # replace all 26 with 167
```

```python
csv1.replace([26,27,25,24],22,inplace=True)  # all the given values is replaced by 22
```

```python
csv1.replace("[A-Za-z]","python",regex=True)  
```

```python
# now we will use ditioary to replace data
```

```python
# when  using dictionary jisko replace krna hai woh key and jisse krna hai usse value 
```

```python
csv1.replace({"year":"ter"},22,regex=True)
```

```python
# we can use ffill and bfill with replace also
```

```python
# using limit we can limit how many values will be replaced
```

```python
csv1.replace(1,method="ffill",limit=2,inplace=True)
```
