```python
import pandas as pd
```

```python
var = pd.read_csv("debt.csv")
```

```python
# we will use dropna and fillna
```

```python
var.dropna()  
# dropping all nan and null values
```

```python
# now we will use axis parameters
```

```python
# o is for row and 1 for column
```

```python
var.dropna(axis=1)  # drops all the missing values from coloumn
```

```python
# to remove entire row and coloumn
```

```python
var.dropna(how='any')  # any removes any null values
```

```python
var.dropna(how="all") # it drop all values if entire row is null not idividual
```

```python
# subset parameters
```

```python
# using subset we can remove null from particular colomns
```

```python
var.dropna(subset=["year"])
```

```python

# use of inplace paramtere which is used to remove null values and make a new dataset
```

```python
var.dropna(inplace=True)
```

```python
# using thresh paramter we can remove null values or entire row accorindg to us
```

```python
var.dropna(thresh=1) # it removes row where a single null value is present
```

```python

# now wew ill fillna to fill missing values
```

```python
var.fillna("pyhon") # it fills all the null values with python values
```

```python

# we ca use dictionary to fill where key beomes coloumn and value becomes data to be filled
```

```python
var.fillna({"year":"system"})
```

```python

# there are two methods to fill missing values bfill and ffil
```

```python
var.fillna(method="ffill")
```

```python

# now to fill along axis and inplace
```

```python
var.fillna("pyhon",axis=1) # it fills all the null values in coloumns with pyhon 
```

```python

# now we will use inplace which create a new dataset
```

```python
var.fillna(12,inplace=True)
```
